```python
df = df[df["Aggregate rating"]> 0]
```

```python
df['Cuisines'].fillna('Unknown', inplace=True)
```

```python
df.columns = (
    df.columns.str.lower().str.replace(" ", "_")
)
```

```python
df['cost_per_person'] = df['average_cost_for_two'] / 2
```

```python
df["rating_weighted"] = df["aggregate_rating"] * np.log1p(df["votes"])
```

```python
df["cuisine_count"] = df["cuisines"].apply(lambda x: len(str(x).split(",")))
```

```python
df["online_delivery"] = df["has_online_delivery"].map({"Yes":1, "No":0})
df["table_booking"] = df["has_table_booking"].map({"Yes":1, "No":0})
```

```python
df["price_category"] = pd.cut(
    df["average_cost_for_two"],
    bins=[0,300,700,1500,5000],
    labels=["Budget","Mid","Premium","Luxury"]
)
```

```python
sns.scatterplot(
    data=df,
    x="cost_per_person",
    y="aggregate_rating",
    alpha=0.4
)
plt.show()
```

```python
sns.boxplot(
    data=df,
    x="online_delivery",
    y="aggregate_rating"
)
plt.show()
```

```python
from collections import Counter

cuisine_series = df["cuisines"].str.split(", ")
top_cuisines = Counter([c for sub in cuisine_series for c in sub])

pd.DataFrame(top_cuisines.most_common(10), columns=["Cuisine","Count"])
```

```python
df["success"] = np.where(
    (df["aggregate_rating"] >= 4) & (df["votes"] >= df["votes"].median()),
    1,
    0
)
```
